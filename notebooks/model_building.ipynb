```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import sqlite3
import statsmodels.api as sm
```

Build a regression model.

```python
conn = sqlite3.connect('sm_app.sqlite')
df = pd.read_sql("select * from restaurant",conn)
df
```

```python
new_df = df[['popularity', 'price', 'rating', 'latitude','longitude','distance','credit_card_True','beer_True','outdoor_seating_True','bikes']]
new_df
```

```python
#Use backward selection and include all variables except bikes
y = new_df['bikes']
X = new_df.drop('bikes', axis=1)
X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
#Distance has the highest p-value so we will remove it
y = new_df['bikes']
X = new_df.drop(['bikes', 'distance'], axis=1)
X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
#Our Adjusted R^2 has increased which means it is a better fit for our model. We will now remove outdoor seating since it has the highest p-value
y = new_df['bikes']
X = new_df.drop(['bikes', 'distance', 'outdoor_seating_True'], axis=1)
X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
y = new_df['bikes']
X = new_df.drop(['bikes', 'distance', 'outdoor_seating_True', 'beer_True'], axis=1)
X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
#Our R^2 has increased again to 0.099, we will continue to remove the highest p-value which is rating.
y = new_df['bikes']
X = new_df.drop(['bikes', 'distance', 'outdoor_seating_True', 'beer_True', 'rating'], axis=1)
X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
y = new_df['bikes']
X = new_df.drop(['bikes', 'distance', 'outdoor_seating_True', 'beer_True', 'rating', 'credit_card_True'], axis=1)
X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
#Our Adj R^2 has decreased from .099 to .098 which means it has a worse fit for our model. We can try again to remove the next highest p-value which is latitude
y = new_df['bikes']
X = new_df.drop(['bikes', 'distance', 'outdoor_seating_True', 'beer_True', 'rating', 'credit_card_True', 'latitude'], axis=1)
X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
#Now we no longer have any p-values above 0.05 so we stop here.
```

Provide model output and an interpretation of the results. 

```python
# I believe this is our best model since it has the highest Adj R^2 value which means it fits out model the best. 
# I chose not to remove credit card and latitude because although it has a high p value > 0.05, the Adj R^2 decreases so
# it is worse for our model.
y = new_df['bikes']
X = new_df.drop(['bikes', 'distance', 'outdoor_seating_True', 'beer_True', 'rating'], axis=1)
X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```
